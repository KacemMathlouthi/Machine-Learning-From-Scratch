```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from LinearRegression import LinearRegression
```

```python
X, y = datasets.make_regression(n_samples = 500, n_features=1, noise = 25, random_state=4)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1234)
```

```python
fig = plt.figure(figsize=(8,6))
plt.scatter(X[:,0], y, color = "b", marker = "o", s = 30)
plt.show()
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
error = mse(y_test, predictions)
print("The MSE Error is : " + str(error))
```

```python
for learning_rate in [0.1, 0.01, 0.001, 0.0001]:
    for n_iters in [100, 1000, 5000, 10000]:
        test_model = LinearRegression(learning_rate=learning_rate, n_iters=n_iters)
        test_model.fit(X_train, y_train)
        predictions = test_model.predict(X_test)
        error = mse(y_test, predictions)
        print("The MSE Error for n_iters = " + str(n_iters) + " and learning rate = " + str(learning_rate) + " is " + str(error))
```
